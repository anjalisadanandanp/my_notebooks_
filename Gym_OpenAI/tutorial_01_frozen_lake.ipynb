```python
import gym
import matplotlib.pyplot as plt
```

### Creating our first Gym environment: Frozen Lake

https://www.gymlibrary.dev/environments/toy_text/frozen_lake/

```python
env = gym.make("FrozenLake-v1", map_name="4x4",render_mode="rgb_array", is_slippery=False)
```

```python
env.reset()     #puts the agent back in the initial state
plt.figure(figsize=(2.5,2.5)) 
plt.axis('off')
plt.imshow(env.render())
```

Number of states in the environment:

```python
print(env.observation_space)
```

Number of actions:

```python
print(env.action_space)
```

0 --> left \
1 --> down \
2 --> right \
3 --> up 

### Transition probability and reward function

env.P[STATE][ACTION] --> [(transition probability, next state, reward, Is terminal state?)]

```python
print(env.P[0][0])
print(env.P[0][1])
print(env.P[0][2])
print(env.P[0][3])
```

```python
print(env.P[10][0])
print(env.P[10][1])
print(env.P[10][2])
print(env.P[10][3])
```

### Taking actions

```python
state = env.reset()
(next_state, reward, done, info, prob) = env.step(2)
print(next_state, reward, done, info, prob)
plt.figure(figsize=(2.5,2.5)) 
plt.axis('off')
plt.imshow(env.render())
```

```python
(next_state, reward, done, info, prob) = env.step(1)
print(next_state, reward, done, info, prob)
plt.figure(figsize=(2.5,2.5)) 
plt.axis('off')
plt.imshow(env.render())
```

### Sampling a random action

```python
state = env.reset()
random_action = env.action_space.sample()
(next_state, reward, done, info, prob) = env.step(random_action)
print(next_state, reward, done, info, prob)
plt.figure(figsize=(2.5,2.5)) 
plt.axis('off')
plt.imshow(env.render())
```

### Generating an episode

```python
state = env.reset()
num_timesteps = 20
for time in range(num_timesteps):
    random_action = env.action_space.sample()
    (next_state, reward, done, info, prob) = env.step(random_action)
    plt.figure(figsize=(2.5,2.5)) 
    plt.axis('off')
    plt.imshow(env.render())
    if done:
        break
```
